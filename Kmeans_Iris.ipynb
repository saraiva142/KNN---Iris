```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
## Aproveitei as bibliotecas da antiga tarefa
```

```python
df = pd.read_csv('iris.data')
```

```python
df.head()
```

```python
df.columns = ['comprimento_sepala', 'largura_sepala', 'comprimento_petala', 'largura_petala', 'classe']
#df = df.rename(columns={'5.1': 'comprimento_sepala', '3.5': 'largura_sepala', '1.4': 'comprimento_petala', '0.2': 'largura_petala', 'iri'})
```

```python
import matplotlib.pyplot as plt

# Define as cores para cada tipo de dado
cores = {'comprimento_sepala': 'red', 'largura_sepala': 'blue', 'comprimento_petala': 'green', 'largura_petala': 'purple'}

# Plota os dados de comprimento e largura da sépala
plt.scatter(df['comprimento_sepala'], df['largura_sepala'], s=32, alpha=0.8, c=cores['comprimento_sepala'], label='Sépala')

# Plota os dados de comprimento e largura da pétala
plt.scatter(df['comprimento_petala'], df['largura_petala'], s=32, alpha=0.8, c=cores['comprimento_petala'], label='Pétala')

# Define os rótulos dos eixos e o título do gráfico
plt.xlabel('Comprimento (cm)')
plt.ylabel('Largura (cm)')
plt.title('Gráfico de Dispersão do Comprimento e Largura da Sépala e Pétala')

# Adiciona uma legenda
plt.legend()

# Remove as bordas superior e direita do gráfico
plt.gca().spines[['top', 'right']].set_visible(False)

# Exibe o gráfico
plt.show()
```

```python
# Transformando os dados em um array numpy
X = np.array(list(zip(df['comprimento_sepala'], df['largura_sepala'], df['comprimento_petala'], df['largura_petala']))) # Fixed: Access columns individually within zip
```

```python
# Aplicando o K-means com 3 clusters
kmeans = KMeans(n_clusters=3).fit(X)
#kmeans.fit(X)
```

```python
# Obtendo os centroids e os labels dos clusters
centroids = kmeans.cluster_centers_
labels = kmeans.labels_
```

```python
# Plotando os clusters
plt.figure(figsize=(10, 6))

plt.scatter(X[:, 0], X[:, 1], c=labels[:X.shape[0]], cmap='rainbow', label='Dados')

plt.scatter(centroids[:, 0], centroids[:, 1], marker='X', s=200, c='black', label='Centroids')


# Plotando os clusters
#plt.figure(figsize=(10, 6))
#plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='rainbow', label='Dados')
plt.scatter(centroids[:, 0], centroids[:, 1], marker='X', s=200, c='black', label='Centroids')

# Definindo o título e rótulos
plt.title('K-means Clustering - Iris_Setosa, Iris_Versicolor e Iris_Virginica')
plt.xlabel('Comprimento (cm)')
plt.ylabel('Largura (cm)')
plt.grid(True)
plt.legend()
#plt.show()

# Contagem de frutas
contagem_setosa = 0
contagem_versicolor = 0
contagem_virginica = 0


species = df['classe'].values

# Exibindo os clusters com as frutas e fazendo a contagem
for i, (classe) in enumerate(X):
    if species[i] == "Iris-setosa":
        iris = "Iris-setosa"
        contagem_setosa += 1
    elif species[i] == "Iris-versicolor":
        iris = "Iris-versicolor"
        contagem_versicolor += 1
    else:
        iris = "Iris-virginica"
        contagem_virginica += 1

    print(f'Comprimento: {df["comprimento_sepala"].iloc[i]},{df["comprimento_petala"].iloc[i]}, Largura: {df["largura_sepala"].iloc[i]}, {df["largura_petala"].iloc[i]} -> {iris}')

# Exibindo a contagem
print("\nContagem de frutas:")
print(f"Setosa: {contagem_setosa}")
print(f"Versicolor: {contagem_versicolor}")
print(f"Virginica: {contagem_virginica}")

plt.show()
```
